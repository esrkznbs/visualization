```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
np.random.seed(613)
x = np.arange(50)
y1 = np.random.randint(0,100,50)
y2 = np.random.randint(0,100,50)
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y1)
ax.plot(x,y2)
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc='right')
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc='lower right',
          frameon = True, #add frame around the legend
          fontsize = 12, #change font size
          ncol = 2, #specift number of columns
          shadow = True )
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc='upper left',
          bbox_to_anchor = (1, 1))
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.text(10, 95, "This value is important!")
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right',prop={"family":"serif","size":20})
ax.text(10, 95, "This value is important!",
       ha='center',
       color='red',
       size= 20)
```

```python
fig, ax = plt.subplots()
ax.axis([0,10,0,10])
ax.text(1,5,'Data:(1,5)',
         transform=ax.transData)
ax.text(0.5,0.1,'Axes:(0.5,0.1)', transform=ax.transAxes)
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.annotate('This is important!', xy=(10, 95), xytext=(20, 94),
            arrowprops=dict(facecolor='black'))
fig.show
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.annotate('This is important!', xy=(10, 95), xytext=(20, 94),
            arrowprops=dict(arrowstyle='<->'), color='green')
fig.show
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.yaxis.set_major_locator(plt.NullLocator())
ax.xaxis.set_major_formatter(plt.NullFormatter())
fig.show
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.yaxis.set_major_locator(plt.MaxNLocator())
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
plt.xticks(rotation=45, ha='right')
```
